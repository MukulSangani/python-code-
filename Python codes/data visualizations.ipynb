```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("F:\download\Pokemon1.csv",header=0,index_col=0)
df.head()
```

```python
df.info()
```

```python
plt.scatter(data=df,x="Attack",y="Defense")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data=df,x="Attack",y="Defense", s=200, c="green",marker="*")
#changes axes ranges
plt.xlim(0,150)
plt.ylim(0,200)
 # add title
plt.title("Relationship Between Attcak and Defense")
 # add xand y
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.grid(True)

# show plot
plt.savefig("test.jpg")
plt.show()
```

```python
# scatter plot with regression line
sns.lmplot(data=df,x="Attack",y="Defense")
```

```python
sns.lmplot(data=df,x="Attack",y="Defense",
           fit_reg=False, # No regression line
          hue="Stage",palette="Set1")   # Color by evolution stage
```

```python
sns.lmplot(data=df.loc[df["Stage"]==3],x="Attack",y="Defense",
           fit_reg=False, # No regression line
          hue="Stage",palette="Set1")   # Color by evolution stage
```

```python
sns.pairplot(df,x_vars=["Attack","Defense","Speed"],y_vars="Total",
            kind="scatter",hue="Stage")
```

```python
# line plot
x=[10,30,20,70]
y=[10,45,10,70]
plt.plot(x,y,c="purple")

plt.xlim(0,100)
plt.ylim(0,100)
plt.title("A simple line plot")
plt.grid(True)
plt.savefig("test.jpg")
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(18,7))
plt.subplot(121)
plt.hist(x="Attack",data=df,bins=10,color="darkblue")
plt.title("Attacks")                                        # bins are number of graph plotted
plt.subplot(122)
plt.hist(x="Defense",data=df,bins=8,color="purple")       
plt.title("Defense")
plt.xticks(np.arange(0,201,10))      # for changing the range for x and y axis
plt.show()
```

```python
# distribution plot
sns.distplot(df.Attack)
```

```python
# count plot (a.k.a Frequency bar plot)
sns.countplot(x="Type 2",data=df,palette="Set2",order=df["Type 2"].value_counts().index)  # inoder to determine the sequnece in 
#    ascending or descending

# rotate x-labels
plt.xticks(rotation=45)
```

```python
import numpy as np
sns.barplot(x="Type 1", y="Attack",data=df,estimator=np.mean) 
plt.xticks(rotation=90)
```
