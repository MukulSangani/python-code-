```python
df1.issue_d=pd.to_datetime(df1.issue_d)
col_name='issue_d'
print (df1[col_name].dtype)
```

```python
split_date="May-2015"
df1_train=df1.loc[df1['issue_d']<=split_date]
df1_train=df1_train.drop(['issue_d'],axis=1)
df1_train.shape
```

```python
df1_test=df1.loc[df1['issue_d']> split_date]
df1_test=df1_test.drop(['issue_d'],axis=1)
df1_test.shape
```

```python
colname=[]
for x in df1.columns:
    if df1[x].dtype=='object':
        colname.append(x)
colname
```

```python
from  sklearn import preprocessing
le=preprocessing.LabelEncoder()
for x in colname:
        df1_train[x]=le.fit_transform(df1_train[x])
```

```python
from  sklearn import preprocessing
le=preprocessing.LabelEncoder()
for x in colname:
        df1_test[x]=le.fit_transform(df1_test[x])
```

```python
X_train=df1_train.values[:,:-1]
Y_train=df1_train.values[:,-1]
```

```python
print(Y_train)
X_test=df1_test.values[:,:-1]
Y_test=df1_test.values[:,-1]

print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X=scaler.transform(X_train)
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_test)
X=scaler.transform(X_test)
print(X_test)
```

```python
from sklearn.linear_model  import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
print(list(zip(Y_test,Y_pred)))  
print(classifier.coef_) 
print(classifier.intercept_)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)
print(classification_report(Y_test,Y_pred))
acc=accuracy_score(Y_test,Y_pred)
print('Accuracy of the model:',acc)
```

```python
Y_pred_prob=classifier.predict_proba(X_test)
print(Y_pred_prob)
```

```python
for a in np.arange(0.4,0.70,0.01):
    predict_mine = np.where(Y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(Y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :",
          cfm[1,0]," , type 1 error:", cfm[0,1]) 
```

```python
from sklearn import metrics

fpr, tpr, z = metrics.roc_curve(Y_test, Y_pred_prob[:,1])
auc = metrics.auc(fpr,tpr)
print(auc)
```

```python
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.show()
```

```python
from sklearn.model_selection import cross_val_score
#running the model using scoring metric as accuracy
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train,
y=Y_train, cv=kfold_cv)
print(kfold_cv_result)
#finding the mean
print(kfold_cv_result.mean())
```
